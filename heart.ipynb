```python
print("Column names in the dataset:", df.columns)
```

```python
chol_mean = df['chol'].mean()
print(f"Mean Cholesterol: {chol_mean}")

age_mean = df['age'].mean()
print(f"Mean Age: {age_mean}")

trtbps_mean = df['trtbps'].mean()
print(f"Mean trtbps: {trtbps_mean}")

thalachh_mean = df['thalachh'].mean()
print(f"Mean thalachh: {thalachh_mean}")

oldpeak_mean = df['oldpeak'].mean()
print(f"Mean oldpeak: {oldpeak_mean}")
```

```python
# Calculate and print median values
chol_median = df['chol'].median()
print(f"Median Cholesterol: {chol_median}")

age_median = df['age'].median()
print(f"Median Age: {age_median}")

trtbps_median = df['trtbps'].median()
print(f"Median trtbps: {trtbps_median}")

thalachh_median = df['thalachh'].median()
print(f"Median thalachh: {thalachh_median}")

oldpeak_median = df['oldpeak'].median()
print(f"Median oldpeak: {oldpeak_median}")
```

```python
# Calculate and print mode values
chol_mode = df['chol'].mode()[0]  # Use the first mode value
print(f"Mode Cholesterol: {chol_mode}")

age_mode = df['age'].mode()[0]
print(f"Mode Age: {age_mode}")

trtbps_mode = df['trtbps'].mode()[0]
print(f"Mode trtbps: {trtbps_mode}")

thalachh_mode = df['thalachh'].mode()[0]
print(f"Mode thalachh: {thalachh_mode}")

oldpeak_mode = df['oldpeak'].mode()[0]
print(f"Mode oldpeak: {oldpeak_mode}")
```
